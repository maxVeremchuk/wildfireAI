```python
train_cols = ['fire_location_latitude', 'fire_location_longitude', 'fire_origin', 'general_cause_desc', 'fire_type', 
              'weather_conditions_over_fire', 'fire_start_date', 'temperature', 'relative_humidity', 'wind_direction', 'wind_speed']
```

```python
wf_df_train = wf_df[train_cols]
```

```python
wf_df_train
```

```python
for name in ['fire_origin', 'fire_type', 'weather_conditions_over_fire', 'wind_direction']:

    one_hot = pd.get_dummies(wf_df_train[name])
    wf_df_train = wf_df_train.join(one_hot)
    wf_df_train = wf_df_train.drop(name,axis = 1)
```

```python
wf_df_train
```

```python
wf_df_train = wf_df_train[~pd.isnull(pd.to_datetime(wf_df_train['fire_start_date'], errors='coerce'))]
```

```python
wf_df_train.fire_start_date = pd.to_datetime(wf_df_train.fire_start_date)
```

```python
wf_df_train['doy'] = wf_df_train.fire_start_date.dt.day_of_year
```

```python
wf_df_train['hour'] = wf_df_train.fire_start_date.dt.hour
```

```python
wf_df_train = wf_df_train.drop('fire_start_date',axis = 1)
```

```python
wf_df_train
```

```python
wf_df_train = wf_df_train.fillna(wf_df_train.mean())
```

```python
wf_df_train.columns
```

```python
norm_cols = ['fire_location_latitude', 'fire_location_longitude', 'temperature','relative_humidity', 'wind_speed', 'doy', 'hour']
```

```python
wf_df_train[norm_cols] = (wf_df_train[norm_cols]-wf_df_train[norm_cols].mean())/wf_df_train[norm_cols].std()
```

```python
definitions = pd.factorize(wf_df_train.general_cause_desc)[1]
```

```python
wf_df_train.general_cause_desc = pd.factorize(wf_df_train.general_cause_desc)[0]
```

```python
wf_df_train
```

```python
y = wf_df_train['general_cause_desc']
X = wf_df_train.drop('general_cause_desc',axis = 1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 21)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
classifier = RandomForestClassifier(n_estimators = 100, criterion = 'entropy', random_state = 42)
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
y_pred
```

```python
reversefactor = dict(zip(range(definitions.shape[0]),definitions))
```

```python
np.sum((y_test - y_pred) == 0)/y_test.shape[0]
```

```python
np.sum((y_test - np.zeros_like(y_test)) == 0)/y_test.shape[0]
```

```python
np.sum(y_test == 0)/y_test.shape[0]
```

```python
y_test_vect = np.vectorize(reversefactor.get)(y_test)
y_pred_vect = np.vectorize(reversefactor.get)(y_pred)
```

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['Lightning', 'Forest Industry', 'Incendiary', 'Resident', 'Recreation',
       'Agriculture Industry', 'Government', 'Power Line Industry',
       'Oil & Gas Industry', 'Undetermined', 'Railroad', 'Other Industry', 'Prescribed Fire', 'Restart'])
```

```python
fig, ax = plt.subplots(figsize=(10,10)) 
disp.plot(xticks_rotation=90, ax=ax)
```

```python
pd.crosstab(y_test_vect, y_pred_vect, rownames=['Actual'], colnames=['Predicted'])
```
